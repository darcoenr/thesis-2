```python
import pandas as pd
import matplotlib.pyplot as plt

from transformers import BertModel
from torch.utils.data import DataLoader
from sklearn.manifold import TSNE

import sys
sys.path.insert(0, '../training/')
import utils
import models
```

### Retrieve the model

```python
MODEL_CHECKPOINT = 'Exscientia/IgBert'
TOKENIZER_CHECKPOINT = None
```

```python
# Get BERT
print('Retrieve transformer:')
print('Model checkpoint: {}'.format(MODEL_CHECKPOINT))
model = BertModel.from_pretrained(MODEL_CHECKPOINT, add_pooling_layer=False)

# Get the tokenizer
print('Retrieve the tokenizer...')
if TOKENIZER_CHECKPOINT is None: TOKENIZER_CHECKPOINT = MODEL_CHECKPOINT
tokenizer = utils.retrieve_tokenizer(TOKENIZER_CHECKPOINT)
```

### Random Pairing

```python
DATA = r'../datasets/new/classificator/train_random/train.csv'
SUBSAMPLE = 4096
FRAC = None
BATCH_SIZE = 32
```

```python
# Get the data
print('Retrieving the datasets...')
data, hyperpar = utils.get_dataset(DATA, subsample=SUBSAMPLE, frac=FRAC)
data_dl = DataLoader(data, batch_size=BATCH_SIZE)
```

```python
# Compute embeddings
print('Compute embeddings...')
embeddings = utils.compute_avg_embeddings(model, tokenizer, data_dl)
```

```python
# Dimensionality reduction
print('Reduce...')
tsne_res = TSNE(2, perplexity=30, n_jobs=-1).fit_transform(embeddings)
```

```python
data_to_plot_random = pd.DataFrame({
        'x': tsne_res[:, 0], 
        'y': tsne_res[:, 1], 
        'class': [label for _, _, label in data],
        'id': [id_seq for id_seq, _, _ in data]})
```

```python
fig, ax = plt.subplots()

ax.scatter(data_to_plot_random[data_to_plot_random['class'] == 0]['x'], 
           data_to_plot_random[data_to_plot_random['class'] == 0]['y'], s=10,
          label='positive')
ax.scatter(data_to_plot_random[data_to_plot_random['class'] == 1]['x'], 
           data_to_plot_random[data_to_plot_random['class'] == 1]['y'], s=10, 
          label='negative')

ax.set_title('Random Pairing')
ax.tick_params(axis='both', which='both', 
                   bottom=False, left=False, labelbottom=False, labelleft=False)
ax.set(xlabel='', ylabel='')
l = ax.legend()


fig.savefig('random-pairing-embeddings.png', bbox_inches='tight')
```

### Germline Pairing

```python
DATA = r'../datasets/only_human-germline_all/splits/train_germline.csv'

SEQ_DATA = r'../datasets/sequences_only_human.csv'
GERM_DATA = r'../datasets/negative_samples_germline_only_human.csv'

SUBSAMPLE = 4096
FRAC = None
BATCH_SIZE = 32
```

```python
# Get the data
print('Retrieving the datasets...')
data, hyperpar = utils.get_dataset(DATA, subsample=SUBSAMPLE, frac=FRAC)
data_dl = DataLoader(data, batch_size=BATCH_SIZE)
```

```python
# Compute embeddings
print('Compute embeddings...')
embeddings = utils.compute_avg_embeddings(model, tokenizer, data_dl)
```

```python
# Dimensionality reduction
print('Reduce...')
tsne_res = TSNE(2, perplexity=30, n_jobs=-1).fit_transform(embeddings)
```

```python
data_to_plot_germline = pd.DataFrame({
        'x': tsne_res[:, 0], 
        'y': tsne_res[:, 1], 
        'class': [label for _, _, label in data],
        'id': [id_seq for id_seq, _, _ in data]})
```

```python
fig, ax = plt.subplots()

ax.scatter(data_to_plot_germline[data_to_plot_germline['class'] == 0]['x'], 
           data_to_plot_germline[data_to_plot_germline['class'] == 0]['y'], s=5,
          label='positive')
ax.scatter(data_to_plot_germline[data_to_plot_germline['class'] == 1]['x'], 
           data_to_plot_germline[data_to_plot_germline['class'] == 1]['y'], s=5, 
          label='negative')

ax.set_title('Germline Pairing')
ax.tick_params(axis='both', which='both', 
                   bottom=False, left=False, labelbottom=False, labelleft=False)
ax.set(xlabel='', ylabel='')
l = ax.legend()

fig.savefig('germline-pairing-embeddings.png', bbox_inches='tight')
```

```python
data_seq = pd.read_csv(SEQ_DATA, index_col=0).drop_duplicates()
data_ger = pd.read_csv(GERM_DATA, index_col=0).drop_duplicates()
```

```python
data_to_plot_germline_pos = data_to_plot_germline[data_to_plot_germline['class'] == 0]
data_to_plot_germline_neg = data_to_plot_germline[data_to_plot_germline['class'] == 1]
```

```python
A = data_to_plot_germline_pos.merge(data_seq, 
                                    left_on='id', 
                                    right_on='pair_id', how='left')[['x', 'y', 'pairing_index']]
B = data_to_plot_germline_neg.merge(data_ger, 
                                    left_on='id', 
                                    right_on='pair_id', how='left')[['x', 'y', 'pairing_index']]

data_to_plot_by_pairing = pd.concat([A, B], axis=0)
```

```python
fig, ax = plt.subplots()

for _, data in data_to_plot_by_pairing.groupby('pairing_index'):
    ax.scatter(data['x'], data['y'], s=5)

ax.set_title('Germline Pairing')
ax.tick_params(axis='both', which='both', 
                   bottom=False, left=False, labelbottom=False, labelleft=False)
ax.set(xlabel='', ylabel='')
```
